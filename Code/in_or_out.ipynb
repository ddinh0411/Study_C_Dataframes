```python
#Returns 0 if point is not in a contour, and 1 if in a contour
def FindPoint(x1, y1, x2,
              y2, x, y) :
    if (x > x1 and x < x2 and
        y > y1 and y < y2) :
        return 1
    else :
        return 0
```

```python
in_outDF = pd.DataFrame({"Image":[],"P_ID":[],"Condition":[], "In_Out":[], "X":[], "Y":[]})
```

```python
# i = 0 #Initializing image index
curr_fix = 0 #Initializing fixation index value
start_hot = 1.0 #Initializing hotspot index value
ans = []

for i in images_navon:
    #print(im)

    #Defining length of fixation loop for each iteration
    im_fix = fixDF.loc[fixDF['Image'] == i]
#     print(im_fix)
    fc = len(im_fix)
    #print(fc)
    #Defining length of hotspot loop for each iteration
    im_hot = hotDF.loc[hotDF['Image'] == i]
#     print(im_hot)
    hc = len(im_hot)


    X_list = im_fix["X"]
    Y_list = im_fix["Y"]
    Participant_ID_list = im_fix["RECORDING_SESSION_LABEL"]
    Cond_list = im_fix["image_type"]
    # fixstart_list = im_fix["CURRENT_FIX_START"]
    # fix_dur_list = im_fix["CURRENT_FIX_DURATION"]
    # Resp_List = im_fix["response"]

    #Fixation loop
    for fix in range(fc):

        X = X_list.iloc[fix]
        Y = Y_list.iloc[fix]


        #Demographic info
        # p_id = im_fix["RECORDING_SESSION_LABEL"][curr_fix]
        # cond = im_fix["filter"][curr_fix]
        # sem = im_fix["type"][curr_fix]
        p_id = Participant_ID_list.iloc[fix]
        cond = Cond_list.iloc[fix]
        # fixstart = fixstart_list.iloc[fix]
        # fix_dur = fix_dur_list.iloc[fix]
        # Resp = Resp_List.iloc[fix]

        curr_fix += 1 #Incrementing value so the index starts @ appropriate value

        #Hotspot loop
        curr_hot = start_hot #Resetting index value so the index starts @ appropriate value

        x1_s = im_hot["X1"]
        x1 = x1_s.to_list()

        x2_s = im_hot['X2']
        x2 = x2_s.to_list()

        y1_s = im_hot['Y1']
        y1 = y1_s.to_list()

        y2_s = im_hot['Y2']
        y2 = y2_s.to_list()

        hot_s = im_hot['Hot_Num']
        hotspot_num = hot_s.to_list()

        cont_s = im_hot['Contour']
        cont = cont_s.to_list()

        hotspot_num = -1.0 # Negative if the person was not looking in a hotspot
        hot_fix_bool = False # This is the variable to determine if the fixation is in a hotspot or not

        for hot in range(hc):

          x1_cord = x1[hot]
          x2_cord = x2[hot]
          y1_cord = y1[hot]
          y2_cord = y2[hot]
          cont_val = cont[hot]

          #Implementing fxn for 1 fixation and each corresponding hotspot
          ans = FindPoint(x1_cord, y1_cord, x2_cord, y2_cord, X, Y) #0

          if(ans == 1):
            hotspot_num = hot + 1
            hot_fix_bool = True
            break

        #ans += ans #Summing all the answers to find whether the point was in a hotspot
        #Appending dataframe with in or out
        #Note: the fixation_num var is not correct
        in_outDF = in_outDF.append({"Image":i, "P_ID":p_id,"Condition":cond, "In_Out":ans, "X": X, "Y":Y}, ignore_index = True)

#         curr_fix = 0
        # print(hotspot_num)

    start_hot += hc #Incrementing value so the index starts @ appropriate value
```
