```python
def average(A,B):
  # Calculate the pairwise Euclidean distances using NumPy
  distances = np.linalg.norm(A.values[:, np.newaxis] - B.values, axis=2)

  # Compute the sum and count of distances
  sum_of_distances = distances.sum()
  count = A.shape[0] * B.shape[0]

  # Calculate the average linkage
  return sum_of_distances / count
```
